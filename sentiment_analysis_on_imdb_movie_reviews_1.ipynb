# Step 1: Understanding the Paper

#### Reading the Paper

**1. Methodology:**

The paper combines unsupervised and supervised techniques to learn word vectors that capture both semantic and sentiment information. Here's a detailed breakdown of the methodology:

- **Unsupervised Learning of Word Vectors:**
  - **Probabilistic Approach:** The paper uses a probabilistic model to learn word vectors. This involves processing a large corpus of text (such as movie reviews) to learn representations of words based on their context in the text. The model aims to capture the semantic relationships between words.
  - **Semantic Term-Document Information:** This part of the model focuses on understanding the context in which words appear, capturing the semantic similarity between words (e.g., "good" and "excellent" might appear in similar contexts).

- **Supervised Refinement of Word Vectors:**
  - **Sentiment Labels:** After learning the initial word vectors in an unsupervised manner, the model refines these vectors using sentiment labels (positive or negative reviews). This supervised step adjusts the word vectors to better capture sentiment information, helping the model distinguish between positive and negative words.
  - **Sentiment Content:** The refinement process ensures that the word vectors incorporate sentiment information, making them more effective for sentiment analysis tasks.

**2. Model Architecture:**

The model architecture described in the paper includes the following components:

- **Pre-trained Word Vectors:**
  - The paper emphasizes the use of word vectors pre-trained using an unsupervised method. These vectors are then used as input features for the sentiment classifier.
  - **Unsupervised Pre-training:** Word vectors are learned from a large text corpus without any labels. This step captures the general semantic relationships between words.
  - **Supervised Refinement:** The pre-trained word vectors are fine-tuned using labeled sentiment data, which helps the model learn the sentiment aspects of words.

- **Simpler Classifier:**
  - The paper uses simpler classifiers like Naive Bayes or Support Vector Machines (SVM) for sentiment classification. The choice of a simpler classifier highlights the effectiveness of the pre-trained word vectors.
  - **Naive Bayes/SVM:** These classifiers take the pre-trained word vectors as input and use them to predict the sentiment of the reviews. The simplicity of these models allows the focus to be on the quality of the word vectors.

**3. Evaluation Metrics:**

The paper evaluates the model using several key metrics to measure its performance:

- **Accuracy:**
  - Measures the proportion of correctly predicted instances out of the total instances.
  - Formula:
    \[
    \text{Accuracy} = \frac{\text{Number of correct predictions}}{\text{Total number of predictions}}
    \]

- **Precision:**
  - Measures the proportion of true positive predictions out of all positive predictions made by the model.
  - Formula:
    \[
    \text{Precision} = \frac{\text{True Positives}}{\text{True Positives} + \text{False Positives}}
    \]

- **Recall:**
  - Measures the proportion of true positive predictions out of all actual positive instances.
  - Formula:
    \[
    \text{Recall} = \frac{\text{True Positives}}{\text{True Positives} + \text{False Negatives}}
    \]

- **F1 Score:**
  - Provides a balance between precision and recall, calculated as the harmonic mean of these two metrics.
  - Formula:
    \[
    \text{F1 Score} = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}
    \]

**3. Evaluation Metrics:**

````python
# Step 3.2.1: Read and Preprocess the Data

import os
import pandas as pd

def read_imdb_data(directory):
    reviews = []
    for subdir, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith(".txt"):
                with open(os.path.join(subdir, file), 'r', encoding='utf-8') as f:
                    reviews.append(f.read())
    return reviews

# Assuming the 'aclImdb' directory is extracted in your Colab environment
train_pos_reviews = read_imdb_data('aclImdb/train/pos')
train_neg_reviews = read_imdb_data('aclImdb/train/neg')
train_reviews = train_pos_reviews + train_neg_reviews

# Tokenize the reviews
tokenized_reviews = [review.split() for review in train_reviews]
````

##    Step 5.2: Define the Function to Convert Reviews to Feature Vectors

1.    **Define the `get_feature_vector` Function:**
      *   This function takes a review, the pre-trained Word2Vec model, and the vector size as inputs.
      *   It splits the review into words and computes the average of the word vectors for all words in the review.
          ```
          # Step 5.2.1: Define the get_feature_vector function

          def get_feature_vector(review, model, vector_size):
            words = review.split()
            feature_vector = np.zeros((vector_size,), dtype="float32")
            num_words = 0
            for word in words:
                if word in model.wv:
                    feature_vector = np.add(feature_vector, model.wv[word])
                    num_words += 1
            if num_words > 0:
                feature_vector = np.divide(feature_vector, num_words)
            return feature_vector
          ```

````python
# Step 5.2.1: Define the get_feature_vector Function

def get_feature_vector(review, model, vector_size):
  words = review.split()
  feature_vector = np.zeros((vector_size,), dtype="float32")
  num_words = 0
  for word in words:
      if word in model.wv:
          feature_vector = np.add(feature_vector, model.wv[word])
          num_words += 1
  if num_words > 0:
      feature_vector = np.divide(feature_vector, num_words)
  return feature_vector
````

````python
# Step 6.2.1: Import Necessary Libraries

from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# Step 6.2.2: Train the Naive Bayes Classifier

nb_model = GaussianNB()
nb_model.fit(X_train, y_train)
````

````python
# Step 6.2.3: Predict and Evaluate

y_pred = nb_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1}')
````

````python
# Plotting Evaluation Metrics
metrics = {'Accuracy': accuracy, 'Precision': precision, 'Recall': recall, 'F1 Score': f1}

plt.figure(figsize=(10, 6))
plt.bar(metrics.keys(), metrics.values(), color=['blue', 'green', 'red', 'purple'])
plt.xlabel('Metrics')
plt.ylabel('Score')
plt.title('Evaluation Metrics')
plt.ylim(0, 1)
plt.savefig('evaluation_metrics.png')
plt.show()
````
